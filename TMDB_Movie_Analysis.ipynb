Defined Functions
¶

```
def scatter_plot(df, x, y, title= 'title', xlabel= 'xlabel', ylabel= 'ylabel', color= 'color'):
    df.plot.scatter(x= x,
                    y= y, 
                    figsize = (8, 8), 
                    color= color
                   )
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def scatter_plot_index(df, x, y, title= 'title', xlabel= 'xlabel', ylabel= 'ylabel'):
    plt.scatter(x= df[x], 
                y= df[y].index
               )
    plt.title(title), 
    plt.xlabel(xlabel), 
    plt.ylabel(ylabel)
    plt.show()
    
def top10_diagram(column, x, y, title= 'title', xlabel= 'xlabel', ylabel= 'ylabel'):
    df.nlargest(10, column).plot(
                                    x = x, 
                                    y= y, 
                                    kind= 'barh', 
                                    title = title, 
                                    xlabel = xlabel, 
                                    ylabel = ylabel, 
                                    figsize = (10, 6),
                                    color = '#2a6099',
                                    );
```

vote average category
¶

Seperate vote average data into four categories using describe function to know the limits

```
df['vote_average'].describe()
```

```
# vote average bigger or equal to 6.7 
vah = df.query('vote_average >= 6.7')
vah_un = vah['vote_average'].unique()
vah_nu = vah['vote_average'].nunique() 
vah_c = vah['vote_average'].count() 

# bigger or equal to 6.2 
vam = df.query('vote_average >= 6.2')
vam_un = vam['vote_average'].unique()
vam_nu = vam['vote_average'].nunique() 
vam_c = vam['vote_average'].count() 

# bigger or equal to 5.7
valw = df.query('vote_average >= 5.7')
valw_un = valw['vote_average'].unique()
valw_nu = valw['vote_average'].nunique() 
valw_c = valw['vote_average'].count() 

# less than 5.7
vavl = df.query('vote_average < 5.7')
vavl_un = vavl['vote_average'].unique()
vavl_nu = vavl['vote_average'].nunique() 
vavl_c = vavl['vote_average'].count()
```

```
# vote average unique
vaun = vah_un, vam_un, valw_un, vavl_un
# vote average category count
vacc = vah_c, vam_c, valw_c, vavl_c
```

```
vaun
```

```
# vote average category names
#vacnm = "High", "Medium", "Low", "Very Low"
```

```
# vote average category names
vacnm = "More than (6.7)", "More than (6.2)", "More than (5.7)", "Less than (5.7)"
```

```
vacnm
```

```
#vote average category counts for each category
vacc
```

```
#the scheduele
#making a dataframe

c1 = vacnm
c2 = vacc
s = c1, c2

vac_sch = pd.DataFrame(list(s), index = ['category', 'count'])
vac_sch = vac_sch.T
```
